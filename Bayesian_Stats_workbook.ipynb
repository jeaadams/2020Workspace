```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def update_posterior(R, N):
    """Args:
            R: number of heads
            N: number of flips
    """
    h_range = np.arange(0,1.1,0.1)
    prior = 1
    likelihood = (h_range**R)*((1-h_range)**(N-R))
    posterior = likelihood*prior
    return posterior
```

```python
h_range = np.arange(0,1.1,0.1)

fig = plt.figure(figsize = (20,15))
ax1 = fig.add_subplot(3,3,1)
ax1.plot(h_range, update_posterior(0,0), label = '0 Flips', color = 'black')
ax1.legend()
ax1.set_ylabel('Posterior')

ax2 = fig.add_subplot(3,3,2)
ax2.plot(h_range, update_posterior(1,1), label = '1 Flip, 1 Heads', color = 'black')
ax2.legend()

ax3 = fig.add_subplot(3,3,3)
ax3.plot(h_range, update_posterior(2,2), label = '2 Flips, 2 Heads', color = 'black')
ax3.legend()

ax4 = fig.add_subplot(3,3,4)
ax4.plot(h_range, update_posterior(2,3), label = '3 Flips, 2 Heads', color = 'black')
ax4.legend()
ax4.set_ylabel('Posterior')

ax5 = fig.add_subplot(3,3,5)
ax5.plot(h_range, update_posterior(2,4), label = '3 Flips, 2 Heads', color = 'black')
ax5.legend()

ax6 = fig.add_subplot(3,3,6)
ax6.plot(h_range, update_posterior(2,8), label = '3 Flips, 2 Heads', color = 'black')
ax6.legend()

ax7 = fig.add_subplot(3,3,7)
ax7.plot(h_range, update_posterior(2,16), label = '3 Flips, 2 Heads', color = 'black')
ax7.legend()
ax7.set_xlabel('Bias-weighting for heads')
ax7.set_ylabel('Posterior')

ax8 = fig.add_subplot(3,3,8)
ax8.plot(h_range, update_posterior(2,32), label = '3 Flips, 2 Heads', color = 'black')
ax8.legend()
ax8.set_xlabel('Bias-weighting for heads')

ax9 = fig.add_subplot(3,3,9)
ax9.plot(h_range, update_posterior(2,64), label = '3 Flips, 2 Heads', color = 'black')
ax9.legend()
ax9.set_xlabel('Bias-weighting for heads')
```
